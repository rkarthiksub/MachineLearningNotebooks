```python
# prompt: # prompt: hypothesis testing and t-test with country data show graph

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Load the data
data = pd.read_csv('Country-data.csv')

# Select the columns of interest
X = data[['country', 'gdpp']]

# Split the data into two groups
group1 = X[X['country'] == 'United States']
group2 = X[X['country'] == 'Canada']

# Perform the t-test
t, p = ttest_ind(group1['gdpp'], group2['gdpp'])

# Print the results
print('t =', t)
print('p =', p)

# Plot the data
plt.scatter(group1['country'], group1['gdpp'])
plt.scatter(group2['country'], group2['gdpp'])
plt.xlabel('Country')
plt.ylabel('GDP per capita')
plt.show()
```

```python
# prompt: example of normalization using country data with graph

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load the data
df = pd.read_csv('Country-data.csv')
print (df)
```

```python

# Normalize the data
scaler = MinMaxScaler()
df[['life_expec', 'gdpp']] = scaler.fit_transform(df[['life_expec', 'gdpp']])

# Plot the data
plt.scatter(df['gdpp'], df['life_expec'])
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
```
